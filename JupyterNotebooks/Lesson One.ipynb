### Hardware vs Software
---

So in the world of computing we have 2ish realms. 

- Hardware
- Software
- Firmware (ish)

In the most basic sense, we have hardware which is our physical. This is made up of servers, storage, memory, mainfraimes.

Inside of those things we also have physical components that go into it. 

Then we have software which is code. This is where we do things like programming, artificial intelligence, machine learning, hacking, cybersecurity, etc.
